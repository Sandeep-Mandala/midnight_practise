```python
bank_df=pd.read_csv("bank.csv",sep=';')
cat_columns=bank_df.select_dtypes(include='object').columns
for i in cat_columns:
    # Step-1: we need to get unique lables
    unique_lables=bank_df[i].unique()
    unique_lables
    
    # Step-2: we need to get a numbers = len unique lables
    list1=[i for i in range(len(unique_lables))]
    
    # Step-3: we need to create a dictionary
    dict1=dict(zip(unique_lables,list1))
    dict1
    
    # I want to apply the map method for all categorical columns
    bank_df[i]=bank_df[i].map(dict1)
bank_df
```
